```python
#write a python program to creat a list of 10 integers. Then perform the following operations-
#Append an element to the list
#Insert an element at a speicific position
#Remove an element from the list
#Pop an element from a specific position

list= input("Enter element separated by space: ").split()
list=[int(x) for x in list]
print("List of 10 integers : ",list)

list.append(11)
print("After appending the list:",list)

list.insert(2, 6)
print("After inserting the list:",list)

list.remove(6)
print("After removing the list:",list)

list.pop(4)
print("After popping the list:",list)
```
