```python
#Write a function to calculate the sum of the squares of a given number

def sum_of_digit_squares(n):
    total = 0
    while n > 0:
        digit = n % 10
        total += digit * digit
        n = n // 10
    return total

number = int(input("Enter a number: "))
result = sum_of_digit_squares(number)
print("Sum of squares of the digits is:", result)

```
