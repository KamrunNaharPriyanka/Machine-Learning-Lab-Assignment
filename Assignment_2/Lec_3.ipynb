```python
#Write a program to calculate the sum of all even numbers from 1 to 100 with for and while loop

#using for loop
sum=0
for i in range(1,100+1):
  if i%2==0:
   sum += i
print(f'Sum of all even numbers from 1 to 100 using for loop is: {sum}')
```

```python
#Using While loop

j=1
sum=0
while j<=100:
  if j%2 == 0:
    sum += j
  j +=1

print(f'Sum of all even numbers from 1 to 100 using while loop is: {sum}')
```
